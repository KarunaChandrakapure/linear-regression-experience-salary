```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df = pd.read_csv('Salary_dataset.csv',index_col =0)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='YearsExperience', y='Salary', data=df)
plt.title('Years of Experience vs Salary')
plt.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[['YearsExperience']],df['Salary'],test_size=0.2,random_state=42)
```

```python
reg = LinearRegression().fit(X_train,y_train)
```

```python
y_pred_test = reg.predict(X_test)
```

```python
print("Slope (Coefficient):", reg.coef_[0])
print("Intercept:", reg.intercept_)
```

```python
print("R² on test data:", r2_score(y_test, y_pred_test))
```

```python
plt.scatter(X_test,y_test,label='Actual')
plt.plot(X_test, y_pred_test, color='red', label='Predicted')
plt.title('Actual vs Predicted')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
